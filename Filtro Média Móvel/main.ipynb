```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = "bolsa.csv"
df_bolsa = pd.read_csv(path)
df_bolsa_fechamento = df_bolsa[['Close']]
```

```python
ordens = [3, 5, 10]

fig, ax = plt.subplots(figsize=(7, 5))

# Plot original
ax.plot(df_bolsa_fechamento['Close'].values[:50], label='Original', color='blue')
ax.set_title('Original')
ax.set_xlabel('Dias')
ax.set_ylabel('Fechamento')
ax.legend()
ax.grid()

# Mostrando o gr√°fico
plt.show()
```

```python
#N√∫mero de Ordens

N = 2


# Calcula a m√©dia m√≥vel com janela de tamanho N
df_bolsa_fechamento[f'MA_{N}'] = df_bolsa_fechamento['Close'].rolling(window=N).mean()

# Configura a figura para plotar o gr√°fico
fig, ax = plt.subplots(figsize=(10, 6))

# Plot original
ax.plot(df_bolsa_fechamento['Close'].values[:50], label='Original', color='blue', linestyle='--', alpha=0.7)

# Plot da m√©dia m√≥vel
ax.plot(df_bolsa_fechamento[f'MA_{N}'].values[:50], label=f'M√©dia M√≥vel Ordem {N}', color='red')

# Configura√ß√µes do gr√°fico
ax.set_title(f'Original e M√©dia M√≥vel de Ordem {N}')
ax.set_xlabel('Dias')
ax.set_ylabel('Fechamento')
ax.legend()
ax.grid()

# Exibe o gr√°fico
plt.show()
```
