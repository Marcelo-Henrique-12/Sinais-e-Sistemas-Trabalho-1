```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, FloatSlider
import ipywidgets as widgets

# Função para criar e atualizar o gráfico
def view_filter(N, b):
    a = (1 - b) / (1 - b**N)
    w = a*b**(np.arange(N-1,-1,-1))
    filtered_close = np.array([(1 / N) * np.sum(close[i - N:i]) for i in range(N, days)])
    exp_filtered_close = [np.sum(np.multiply(close[i:i + N], w)) for i in range(days - N)]
    idx_filtered_close = np.arange(N + 1, days + 1)

    plt.figure(figsize=(10, 6))
    plt.plot(idx_filtered_close, close[N:], '-ro', label='Dados')
    plt.plot(idx_filtered_close, filtered_close, '-bo', label='Dados Filtrados (MA)')
    plt.plot(idx_filtered_close, exp_filtered_close, '-go', label='Dados Filtrados (EWMA)')
    plt.title('Dados')
    plt.xlabel('Dia ($n$)')
    plt.ylabel('Valor')
    plt.legend()
    plt.show()

# Exemplo de uso
days = 100  # Substitua pelo valor desejado
close = np.random.randn(days).cumsum()  # Substitua pelos seus dados reais

# Sliders para interatividade
slider_N = IntSlider(value=10, min=1, max=days, step=1, description='N')
slider_b = FloatSlider(value=0.9, min=0.01, max=0.99, step=0.01, description='b')

# Interatividade com ipywidgets
interact(view_filter, N=slider_N, b=slider_b);
```
